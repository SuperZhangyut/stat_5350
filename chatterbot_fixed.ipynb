```python
import logging

# 启用调试日志
# logging.basicConfig(level=logging.INFO)
# Create object of ChatBot class with Storage Adapter
bot = ChatBot(
    'ServiceAgent',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite:///database.sqlite3'
)
```

```python
# Create object of ChatBot class with Logic Adapter
bot = ChatBot(
    'ServiceAgent',  
    logic_adapters=[
        'chatterbot.logic.BestMatch',
        'chatterbot.logic.TimeLogicAdapter'],
)
```

```python
from nltk.tokenize import PunktSentenceTokenizer
import os

# 手动加载 punkt 数据
punkt_path = os.path.join('d:\\YT\\YTpaper\\insiderinformedcyber\\GPU\\myenv\\lib\\nltk_data', 'tokenizers', 'punkt', 'english.pickle')

with open(punkt_path, 'rb') as f:
    tokenizer = PunktSentenceTokenizer(f)

text = "Hello there! How are you? I hope you are doing well."
sentences = tokenizer.tokenize(text)
print(sentences)
```

```python
# https://chatterbot.readthedocs.io/en/latest/training.html#training-with-corpus-data

# Inport ListTrainer
from chatterbot.trainers import ListTrainer

trainer = ListTrainer(bot)

trainer.train([
'Hi',
'Hello',
'I need your assistance regarding my order',
'What is your concern regarding this order?',
'I have a complaint.',
'Please elaborate it.',
'How long it will take to receive an order ?',
'An order takes 3-5 Business days to get delivered.',
'How can I return the package ?',
'You must return within 7 days once you received it.',
'Any delivery fee I need to pay for ?',
'No, you do not need to pay.',
'Ok, thanks.',
'No Problem! Have a Good Day!'
])
```

```python
# Get a response to the input text 'I would like to book a flight.'
response = bot.get_response('I have a complaint.')

print("Bot Response:", response)
```

```python
import sys
name=input("Enter Your Name: ")
print("Name entered:", name)
print("Welcome to the Bot Service! Let me know how can I help you?")
sys.stdout.flush()
while True:
    request=input(name+':')
    if request=='Bye' or request =='bye':
        print('Bot: Bye')
        break
    else:
        response=bot.get_response(request)
        print('Bot:',response)
```
